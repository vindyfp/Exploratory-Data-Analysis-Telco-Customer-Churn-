```python
#3. Perform deep-dive exploration
#Ask minimum 2 questions
#At least 1 of them should involve group-by aggregation!

#a. Berapa average tenure, average MonthlyCharges, dan totalcharges berdasarkan group gender?
#b. TOP 10 customer with the highest TotalCharges
#c. Berapa total customer dan average TotalCharges berdasarkan Type Contract
#d. Berapa total user churn dan non churn
```

```python
df.info()
```

```python
df.describe()
```

```python
#a. Berapa average tenure, average MonthlyCharges, dan Total charges berdasarkan gender?

df.groupby(['gender']).agg(
    avg_tenur = ('tenure', 'mean'),
    avg_MonthlyCharges = ('MonthlyCharges', 'mean'),
    total_Charges = ('TotalCharges', 'sum')
).reset_index()
```

```python
df.groupby(['SeniorCitizen']).agg(
    avg_tenur = ('tenure', 'mean'),
    avg_MonthlyCharges = ('MonthlyCharges', 'mean'),
    total_Charges = ('TotalCharges', 'sum')
).reset_index()
```

```python
#b. TOP 10 customer with the highest TotalCharges
top10_customer = (df
                 .groupby('customerID')
                 .agg(Total_Charge = ('TotalCharges', 'sum'))
                 .reset_index()
                 .sort_values('Total_Charge', ascending=False)
                 .head(10))
top10_customer
```

```python
#c. Berapa total customer dan average TotalCharges berdasarkan Type Contract

data_df = df.groupby(['Contract']).agg(
    total_Cust = ('customerID', 'count'),
    avg_TotalCharges = ('TotalCharges', 'mean')
).reset_index()

data_df
```

```python
fig, ax = plt.subplots(1,2,figsize=(14,7))

sns.barplot(data=data_df, x='Contract', y='total_Cust', ax=ax[0])

sns.barplot(data=data_df, x='Contract', y='avg_TotalCharges', ax=ax[1])
```

```python
#d. Berapa total user churn dan non churn

df['User_Churn'] = df['Churn']
Total_UserChurn = df.groupby(['Churn'], as_index=False)[['User_Churn']].count()
Total_UserChurn
```

```python
UserChurn = Total_UserChurn.rename(columns={'Churn':'Churn Status','User_Churn':'Total_User'})
UserChurn
```

```python
fig, ax = plt.subplots(figsize=(4,6))

sns.barplot(x ='Churn Status', y='Total_User',data=UserChurn,ax=ax, palette=['#005faf','#ff8700'])
```

```python
TotalCharges_median = df['TotalCharges'].median()
print(TotalCharges_median)
```

```python
df['CustomerValue'] = df['TotalCharges'].apply(lambda x: 'High Value' if x >= TotalCharges_median else 'Low Value')
df.head()
```

```python
df_CustomerValue = df.groupby(['CustomerValue']).agg(
    total_Cust = ('CustomerValue', 'count'),
).reset_index()

df_CustomerValue
```
